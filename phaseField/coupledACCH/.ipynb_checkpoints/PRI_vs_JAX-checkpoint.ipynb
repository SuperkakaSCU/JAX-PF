```python
import meshio
import os

jax_file = os.path.join(os.getcwd(), 'explicit_fem/output/vtk/p_000000.vtu')
prisms_file = os.path.join(os.getcwd(), 'PRISMS_PF/Comparison/solution-000000.vtu')

# 读取 JAX 生成的初始条件
mesh_jax = meshio.read(jax_file)

# 读取 PRISMS-PF 生成的初始条件
mesh_prisms = meshio.read(prisms_file)
points = mesh_prisms.points  # (N, 3) or (N, 2) depending on dimension
x_min, y_min = np.min(points[:, 0]), np.min(points[:, 1])
x_max, y_max = np.max(points[:, 0]), np.max(points[:, 1])

domain_size = (x_max - x_min, y_max - y_min)

print("Domain size (Lx, Ly):", domain_size)
print("Cell types:", mesh_prisms.cells_dict.keys())
print("Number of cells of each type:")
for cell_type, cell_array in mesh_prisms.cells_dict.items():
    print(f"  {cell_type}: {len(cell_array)}")

print(points.shape[0])
```

```python
c_jax = mesh_jax.point_data["c"]
n_jax = mesh_jax.point_data["n"]

print(c_jax.shape)
print(n_jax.shape)

# 提取 PRISMS-PF 数据
c_prisms = mesh_prisms.point_data["c"]
n_prisms = mesh_prisms.point_data["n"]

print(c_prisms.shape)
print(n_prisms.shape)
```

```python
import numpy as np

# 计算最大绝对差异
c_diff = np.max(np.abs(c_jax - c_prisms))
n_diff = np.max(np.abs(n_jax - n_prisms))

print(f"最大 c 差异: {c_diff}")
print(f"最大 n 差异: {n_diff}")
```

```python
import matplotlib.pyplot as plt

# 可视化 JAX 的 c 字段
plt.figure()
plt.tricontourf(mesh_jax.points[:, 0], mesh_jax.points[:, 1], c_jax.flatten(), levels=100)
plt.title("JAX - c")
plt.colorbar()

# 可视化 PRISMS-PF 的 c 字段
plt.figure()
plt.tricontourf(mesh_prisms.points[:, 0], mesh_prisms.points[:, 1], c_prisms.flatten(), levels=100)
plt.title("PRISMS-PF - c")
plt.colorbar()

plt.show()
```

```python
# 差值场
dc = c_jax - c_prisms
dn = n_jax - n_prisms

# 网格点坐标
x, y = mesh_jax.points[:, 0], mesh_jax.points[:, 1]

print("x shape:", x.shape)
print("y shape:", y.shape)
print("dc shape:", dc.shape)
print("dn shape:", dn.shape)

# 绘图 Δc
plt.figure(figsize=(6, 5))
contour1 = plt.tricontourf(x, y, dc, levels=100, cmap='coolwarm')
plt.colorbar(contour1, label=r'$\Delta c = c_\mathrm{JAX} - c_\mathrm{PRISMS}$')
plt.title("Difference Field: Δc")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.tight_layout()

# 绘图 Δn
plt.figure(figsize=(6, 5))
contour2 = plt.tricontourf(x, y, dn, levels=100, cmap='coolwarm')
plt.colorbar(contour2, label=r'$\Delta n = n_\mathrm{JAX} - n_\mathrm{PRISMS}$')
plt.title("Difference Field: Δn")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.tight_layout()

plt.show()
```
